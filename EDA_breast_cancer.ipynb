```python
cols=[ 'worst_radius', 'worst_texture', 'worst_perimeter', 'worst_area',
       'worst_smoothness', 'worst_compactness', 'worst_concavity',
       'worst_concave_points', 'worst_symmetry', 'worst_fractal_dimension',
       'mean_perimeter', 'mean_area']
df.drop(columns=cols,axis=1,inplace=True)
```

```python
df.shape
```

```python
X=df.drop('target',axis=1)
y=df.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
prediction1 = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm = confusion_matrix(y_test,prediction1)
```

```python
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
accuracy_score(y_test,prediction1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB 
```

```python
dtc = DecisionTreeClassifier()
rfc = RandomForestClassifier()
svc = SVC()
knn = KNeighborsClassifier()
nb = GaussianNB()
```

```python
dtc.fit(X_train,y_train)
rfc.fit(X_train,y_train)
svc.fit(X_train,y_train)
knn.fit(X_train,y_train)
nb.fit(X_train,y_train)
```

```python
prediction2 = dtc.predict(X_test)
prediction3 = rfc.predict(X_test)
prediction4 = svc.predict(X_test)
prediction5 = knn.predict(X_test)
prediction6 = nb.predict(X_test)
```

```python
cm2 = confusion_matrix(y_test,prediction2)
cm3 = confusion_matrix(y_test,prediction3)
cm4 = confusion_matrix(y_test,prediction4)
cm5 = confusion_matrix(y_test,prediction5)
cm6 = confusion_matrix(y_test,prediction6)
```

```python
sns.heatmap(cm2,annot=True)
```

```python
sns.heatmap(cm3,annot=True)
```

```python
sns.heatmap(cm4,annot=True)
```

```python
sns.heatmap(cm5,annot=True)
```

```python
sns.heatmap(cm6,annot=True)
```

```python
print("LogisticRegression Accuracy : {}".format(accuracy_score(y_test,prediction1)))
print("DecisionTree Accuracy : {}".format(accuracy_score(y_test,prediction2)))
print("RandomForest Accuracy : {}".format(accuracy_score(y_test,prediction3)))
print("SVM Accuracy : {}".format(accuracy_score(y_test,prediction4)))
print("KNN Accuracy : {}".format(accuracy_score(y_test,prediction5)))
print("GaussianNB Accuracy : {}".format(accuracy_score(y_test,prediction6)))
```

# As We can see Random Forest Classifier is giving highest accuracy of 95.32% followed by DecsionTree and  LogisticRegression with accuracy score of 93.56% and 90.64% respectively.
